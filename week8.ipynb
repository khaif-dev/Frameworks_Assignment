```python
#importing the necessary libraries
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the data but limiting it to 5000 rows because its large
df= pd.read_csv("metadata.csv", nrows=5000)
```

```python
#previwing the data
df.head()
```

```python
#checking the data frame size
print(df.shape)
```

```python
#checking for the data type of each column
print(df.dtypes)
```

```python
#gettng more information on the data
print(df.info)
```

```python
#checking for missing values
print(df.isnull().sum())
```

```python
# generating basic statistics
print(df.describe())
```

```python
# finding columns with most missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

print("Columns with missing values:")
print(missing)
```

```python
#dropping all columns where all 5000rows have missing values
df_cleaned = df.dropna(axis=1, how="all")

#comparing the original data frame with the cleaned one
print("Original shape:", df.shape)
print("Cleaned shape:", df_cleaned.shape)
```

```python
# converting publish time to date time format
df_cleaned = df_cleaned.copy()
df_cleaned['publish_time'] = pd.to_datetime(
    df_cleaned['publish_time'], errors='coerce'
)
```

```python
#getting the year of publsh
df_cleaned['pub_year'] = df_cleaned['publish_time'].dt.year
```

```python
#creating the word count column and populating it
df_cleaned['abstract_word_count'] = (
    df_cleaned['abstract']
    .astype(str)          
    .apply(lambda x: len(x.split()))
)
```

```python
# checking for the cleaning
print(df_cleaned[['publish_time', 'pub_year', 'abstract_word_count']].head())
```

```python
#counting the numbers of papers per year
papers_per_year = df_cleaned['pub_year'].value_counts().sort_index()
print(papers_per_year)

# ploting the number of paper per year
plt.figure(figsize=(10,5))
papers_per_year.plot(kind='bar', color='green')
plt.title("Papers per Year")
plt.xlabel("Year")
plt.ylabel("Number of Papers")
plt.show()
```

```python
# checking for top journals publishing COVID-19 research
top_journals = df_cleaned['journal'].value_counts().head(10)
print(top_journals)
```

```python
#graph for top publishing jornals
top_journals = df_cleaned['journal'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_journals.plot(kind='bar', color='pink')
plt.title("Top 10 Journals Publishing COVID-19 Research")
plt.xlabel("Journal")
plt.ylabel("Number of Papers")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Finding most frequent words in titles (using simple word frequency)

from collections import Counter
import re

# joining all titles to one big string
all_titles = " ".join(df_cleaned['title'].dropna().astype(str))

# cleaning and spliting words
words = re.findall(r'\b\w+\b', all_titles.lower())

# counting frequency of words
word_freq = Counter(words)

# geting top 10 most common words
common_words = word_freq.most_common(10)
print(common_words)

```

```python
#distribution of paper count by source
papers_by_source = df_cleaned['source_x'].value_counts().head(10)

plt.figure(figsize=(10,5))
papers_by_source.plot(kind='bar', color='lightgreen')
plt.title("Top Sources of Papers")
plt.xlabel("Source")
plt.ylabel("Number of Papers")
plt.xticks(rotation=45, ha='right')
plt.show()
```
